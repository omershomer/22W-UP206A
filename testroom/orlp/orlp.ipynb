# Mapping Past LWCF Grants Outdoor Recreation Legacy Partnership

- https://lwcf.tplgis.org/mappast/

The ORLP layer for their interactive map can be found here:

- https://server5.tplgis.org/arcgis5/rest/services/National/LWCF_Data/FeatureServer/4

Querying where program = 'ORLP' and exporting as a geojson file (included in this folder)

```python
import geopandas as gpd
import contextily as ctx

# to give more power to your figures (plots)
import matplotlib.pyplot as plt
```

```python
orlp = gpd.read_file('orlp.geojson')
```

```python
orlp.head()
```

```python
orlp = orlp.to_crs(epsg=3857)
```

```python
fig,ax = plt.subplots(figsize=(12,12))

orlp.plot(ax=ax,markersize=100,alpha=0.5,linewidth=2,edgecolor='white',color='red')

# get rid of the axis
ax.axis('off');

# basemap from carto that has a dark background (easier to see)
ctx.add_basemap(ax)
```
